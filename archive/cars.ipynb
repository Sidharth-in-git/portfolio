```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Cars.csv", encoding='latin1')
df.head(40)
```

```python
df.columns
```

```python
plt.hist(df['Company Names'].head(40), bins=20, edgecolor='black')
plt.tight_layout()
plt.xticks(rotation=90, fontsize=10)
plt.xlabel('BRANDS', fontsize=20, font='Times New Roman')
plt.ylabel('CARS SOLD', font='Times New Roman', fontsize=20)
plt.show()

```

```python
count = df['Company Names'].value_counts().head(20)
plt.figure(figsize=(8,8))
plt.bar(count.index, count.values, color='green', edgecolor='black')
plt.xlabel('COMPANIES', font='Times New Roman', fontsize=15)
plt.ylabel('NUMBER OF CARS', font='Times New Roman', fontsize=15)
plt.xticks(rotation=90)
plt.grid(color='black')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['Cars Names'].head(20), df['Engines'].head(20), marker="o", color='green')
plt.xticks(rotation=90, fontsize=10)
plt.grid(True)
plt.xlabel('NAMES')
plt.ylabel('Engines')
plt.show()
```

```python
df.dtypes
```

```python
sns.histplot(df['HorsePower'].head(20), bins=20)
plt.xticks(rotation= 90, fontsize= 8)
```

```python
df['HorsePower'] = df['HorsePower'].str.replace('[^0-9.-]', '', regex=True)
df['HorsePower'] = pd.to_numeric(df['HorsePower'], errors='coerce')
```

```python

sns.barplot(x='Engines', y= 'HorsePower',  data=df.head(20), orient='v')

```

```python
plt.Figure(figsize=(8,8))
sns.boxplot(x='Company Names', y='Cars Names',data=df.head(45), hue='Seats')
plt.legend(loc='upper left')
plt.tight_layout()
plt.xticks(rotation=90)
plt.yticks(fontsize=9)
```
